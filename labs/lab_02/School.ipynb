### Converting all categorical data into integers

```python
# convert 't' and 'f' into 1 and 0
def t_f(x):
    school_charter = {}
    school_charter ['f'] = 0
    school_charter ['t'] = 1
    
    return school_charter[x]

df['school_charter'] = df['school_charter'].map(t_f)
```

```python
def t_f(x):
    school_magnet = {}
    school_magnet ['f'] = 0
    school_magnet ['t'] = 1
    
    return school_magnet[x]

df['school_magnet'] = df['school_magnet'].map(t_f)
```

```python
def t_f(x):
    school_year_round = {}
    school_year_round ['f'] = 0
    school_year_round ['t'] = 1
    
    return school_year_round [x]

df['school_year_round'] = df['school_year_round'].map(t_f)
```

```python
def t_f(x):
    school_nlns = {}
    school_nlns ['f'] = 0
    school_nlns ['t'] = 1
    
    return school_nlns[x]

df['school_nlns'] = df['school_nlns'].map(t_f)
```

```python
def t_f(x):
    school_kipp = {}
    school_kipp ['f'] = 0
    school_kipp ['t'] = 1
    
    return school_kipp [x]

df['school_kipp'] = df['school_kipp'].map(t_f)
```

```python
def t_f(x):
    school_charter_ready_promise = {}
    school_charter_ready_promise ['f'] = 0
    school_charter_ready_promise ['t'] = 1
    
    return school_charter_ready_promise[x]

df['school_charter_ready_promise'] = df['school_charter_ready_promise'].map(t_f)
```

```python
def t_f(x):
    teacher_teach_for_america = {}
    teacher_teach_for_america ['f'] = 0
    teacher_teach_for_america ['t'] = 1
    
    return teacher_teach_for_america[x]

df['teacher_teach_for_america'] = df['teacher_teach_for_america'].map(t_f)
```

```python
def t_f(x):
    teacher_ny_teaching_fellow = {}
    teacher_ny_teaching_fellow ['f'] = 0
    teacher_ny_teaching_fellow ['t'] = 1
    
    return teacher_ny_teaching_fellow[x]

df['teacher_ny_teaching_fellow'] = df['teacher_ny_teaching_fellow'].map(t_f)
```

```python
def t_f(x):
    eligible_double_your_impact_match = {}
    eligible_double_your_impact_match ['f'] = 0
    eligible_double_your_impact_match ['t'] = 1
    
    return eligible_double_your_impact_match[x]

df['eligible_double_your_impact_match'] = df['eligible_double_your_impact_match'].map(t_f)
```

```python
def t_f(x):
    eligible_almost_home_match = {}
    eligible_almost_home_match ['f'] = 0
    eligible_almost_home_match ['t'] = 1
    
    return eligible_almost_home_match[x]

df['eligible_almost_home_match'] = df['eligible_almost_home_match'].map(t_f)
```

```python
%%time
# convert poverty levels into 1,2,3,4 from low to high
def pov_num(numb):
    
    poverty_level = {}
    poverty_level ['low poverty'] = 1
    poverty_level ['moderate poverty'] = 2
    poverty_level ['high poverty'] = 3
    poverty_level ['highest poverty'] = 4
    
    return poverty_level[numb]

df['poverty_number'] = df['poverty_level'].map(pov_num)
```

```python
%%time
# convert poverty levels into 1,2,3,4 from low to high
def fund_number(numb):
    
    funding_status = {}
    funding_status ['completed'] = 1
    funding_status ['live'] = 2
    funding_status ['reallocated'] = 3
    funding_status ['expired'] = 4
    
    return funding_status[numb]

df['funding_number'] = df['funding_status'].map(fund_number)
```
